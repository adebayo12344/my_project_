```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 

```

```python
ds=pd.read_csv("Algerian_forest_fires_dataset.csv")
```

```python
ds.head()
```

```python
ds.info()
```

## Data Cleaning

```python
# Checking for missing values
ds[ds.isnull().any(axis=1)]
```

```python
ds.loc[:122,"Region"]=0
ds.loc[122:,"Region"]=1
```

```python
ds.head()
```

```python
ds.info()
```

```python
ds[['Region']]=ds[['Region']].astype(int)
```

```python
ds.isnull().sum()
```

## Remove the null value

```python
## Remove null values
ds=ds.dropna().reset_index(drop=True)
```

```python
ds.head()
```

```python
ds.isnull().sum()
```

```python
ds.iloc[[122]]
```

```python
## Removal of 122 row contaning header for the second region
ds=ds.drop(122).reset_index(drop=True)
```

```python
ds.iloc[[122]]
```

## Fixing columns name of "spaces" and others typo-errors

```python
ds.columns=ds.columns.str.strip()
ds.columns
```

### Change some columns as interger datatype

```python
ds[['month', 'day','year', 'Temperature','RH','Ws']]=ds[['month', 'day','year', 'Temperature','RH','Ws']].astype(int)
ds.info()
```

### Changing the other columns to float

```python
objects=[features for features in ds.columns if ds[features].dtypes=='O']
objects
```

```python
for i in objects:
    if i!='Classes':
     ds[i]=ds[i].astype(float)
```

```python
ds.info()
```

```python
ds.describe()
```

```python
ds.head()
```

## Save the cleaned dataset into .CSV format

```python
ds.to_csv('Cleaned_Algerian_forest_fire_dataset.csv',index=False)
```

## Monthly Fire Analysis

```python
ds['Classes']=np.where(ds['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
dstemp=ds.loc[ds['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=ds)
plt.ylabel('Number Of Fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title('Fire Analysis of Sidibel Region',weight="bold")
```

```python
dstemp=ds.loc[ds['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=ds)
plt.ylabel('Number Of Fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title('Fire Analysis of Bejaia Region',weight="bold")
```
